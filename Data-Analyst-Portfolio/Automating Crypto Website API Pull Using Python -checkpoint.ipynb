```python
import os
import json
import pandas as pd
from requests import Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
from time import sleep

# Initialize your tracking DataFrame as an empty instance
df = pd.DataFrame()
```

```python
def api_runner():
    global df
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest' 
    parameters = {
        'start': '1', 
        'limit': '10', 
        'convert': 'USD'
    }
    headers = {
        'Accepts': 'application/json',
        'X-CMC_PRO_API_KEY': '0ad53085-1cb2-4eb8-ad9e-3ffbd7e56509',
    }

    session = Session()
    session.headers.update(headers)

    try:
        response = session.get(url, params=parameters)
        data = json.loads(response.text)
        
        # 1. Normalize the incoming JSON data batch cleanly
        df2 = pd.json_normalize(data['data'])
        df2['Timestamp'] = pd.Timestamp.now()  # Dynamic, up-to-date timestamp
        
        # 2. Append directly to your global tracking DataFrame
        df = pd.concat([df, df2], ignore_index=True)
        
        csv_path = r'C:\Users\DELL\OneDrive\Pictures\Music\Videos\download\python\API.csv'
        
        # 3. FIX: Save using true/false logic so your CSV formatting stays pristine
        if not os.path.isfile(csv_path):
            df2.to_csv(csv_path, header=True, index=False)
        else:
            df2.to_csv(csv_path, mode='a', header=False, index=False)
            
    except (ConnectionError, Timeout, TooManyRedirects) as e:
        print(e)
```

```python
for i in range(333):
    api_runner()
    print("API Runner completed successfully")
    sleep(30)  # Standard 30-second delay between runs
```
