```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
import datetime
import time
import smtplib
```

```python
URL = "https://www.amazon.in/Unchartered-Legacy-Thieves-Collection-PlayStation/dp/B09N9MBVD1"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}
```

```python
page = requests.get(URL, headers=headers)

soup = BeautifulSoup(page.content, "html.parser")

title = soup.find(id="productTitle").get_text().strip()

price = soup.find("span", class_="a-price-whole").get_text().strip()

price = int(price.replace(",", "").rstrip("."))
print("Product:", title)
print("Price:", price)
```

```python
today = datetime.date.today()

print(today)
```

```python
header = ["Title", "Price", "Date"]

data = [title, price, today]

with open("AmazonScraperDataset.csv", "w", newline="", encoding="UTF8") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)

print("CSV Created Successfully!")
```
